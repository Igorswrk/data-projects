# Meteor Challenge

#### Initialization and assignments

```python
import cv2
import numpy as np
from skimage.io import imshow

img = cv2.imread('./meteor_challenge_01.png') # reading in bgr format
water_level = np.zeros_like(img)

count_number_meteor = 0
count_number_stars = 0
count_number_meteor_falling_water = 0

lin, col, _ = img.shape

print(img.shape)

view_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
imshow(view_img)
```

#### Question 1, 2

```python
for i in range(lin):
    for j in range(col):

        (b, g, r) = img[i, j] # Accessing the values of the r, g, b layers for the analyzed pixel

        PURE_RED = (b == 0 and g == 0 and r == 255) # Meteors
        PURE_BLUE = (b == 255 and g == 0 and r == 0) # Water
        PURE_WHITE = (b == 255 and g == 255 and r == 255) # Stars
        PURE_BLACK = (b == 0 and g == 0 and r == 0) # Ground

        if PURE_RED:
            count_number_meteor +=  1
        elif PURE_WHITE :
            count_number_stars += 1
        elif PURE_BLUE:
            water_level[:, j] = 255 # Selection of vertical water coverage area

print(f"Number of meteors: {count_number_meteor}")
print(f"Number of stars: {count_number_stars}")
```

#### Question 3

```python
for i in range(lin):
    for j in range(col):
        (b, g, r) = img[i, j]
        (b2, g2, r2) = water_level[i, j]
        if (b == 0 and g == 0 and r == 255) and (b2 == 255 and g2 == 255 and r2 == 255):
            count_number_meteor_falling_water += 1

print(f"Meteors falling water: {count_number_meteor_falling_water}")
```
